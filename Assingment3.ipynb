# NumPy Operations

### 1. Create a NumPy array of random numbers and normalize the data using min-max scaling.

```python
import numpy as np
random_array = np.random.rand(10)
print("Original array:", random_array)
min_val = random_array.min()
max_val = random_array.max()
normalized_array = (random_array - min_val) / (max_val - min_val)
print("Normalized array:", normalized_array)
```

### 2. Generate a 5x5 matrix with random integers and replace all occurrences of a specific value with another.

```python
matrix = np.random.randint(0, 10, size=(5, 5))
print("Original 5x5 matrix:\n", matrix)
value_to_replace = 5
new_value = 99
replaced_matrix = np.where(matrix == value_to_replace, new_value, matrix)
print(f"\nMatrix after replacing {value_to_replace} with {new_value}:\n", replaced_matrix)
```

### 3. Create two NumPy arrays and perform element-wise addition, subtraction, multiplication, and division.

```python
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])
print("Array 1:", array1)
print("Array 2:", array2)
addition_result = array1 + array2
print("\nAddition:", addition_result)
subtraction_result = array1 - array2
print("Subtraction:", subtraction_result)
multiplication_result = array1 * array2
print("Multiplication:", multiplication_result)
division_result = array1 / array2
print("Division:", division_result)
```

### 4. Solve the system of linear equations: $3x + 4y = 7$ and $5x + 2y = 8$ using NumPy.

```python
A = np.array([[3, 4], [5, 2]])
B = np.array([7, 8])
print("Coefficient matrix A:\n", A)
print("Constant vector B:\n", B)
solution = np.linalg.solve(A, B)
print("Solution (x, y):", solution)
```

### 10. Implement a function using NumPy that returns the inverse of a given matrix.

```python
def matrix_inverse(matrix):
    matrix = np.array(matrix)
    if matrix.shape[0] != matrix.shape[1]:
        raise ValueError("Matrix must be square to find inverse.")
    det = np.linalg.det(matrix)
    if det == 0:
        raise ValueError("Matrix is singular and cannot be inverted.")
    inverse = np.linalg.inv(matrix)
    return inverse
matrix = [[1, 2],
          [3, 4]]
inv_matrix = matrix_inverse(matrix)
print(inv_matrix)
```

### 14. Merge two DataFrames on a common key and perform an inner, outer, left, and right join.

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['A', 'B', 'C']
})
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Salary': [20000, 30000, 40000]
})
inner_join = pd.merge(df1, df2, on='ID', how='inner')
left_join = pd.merge(df1, df2, on='ID', how='left')
right_join = pd.merge(df1, df2, on='ID', how='right')
outer_join = pd.merge(df1, df2, on='ID', how='outer')
print("Inner_join:\n",inner_join)
print("\nLeft_join:\n",left_join)
print("\nRight_join:\n",right_join)
print("\nOuter_join:\n",outer_join)
```

### 23. Normalize a dataset using StandardScaler from scikit-learn and explain the difference between min-max scaling and standard scaling.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
df_scaling = df_cleaned[['Age', 'Fare']].copy()
df_scaling['Age'] = df_scaling['Age'].fillna(df_scaling['Age'].median())
scaler_std = StandardScaler()
df_scaling['Age_Standardized'] = scaler_std.fit_transform(df_scaling[['Age']])
scaler_minmax = MinMaxScaler()
df_scaling['Age_Normalized'] = scaler_minmax.fit_transform(df_scaling[['Age']])
print("Comparison of Scaling Methods:")
display(df_scaling[['Age', 'Age_Standardized', 'Age_Normalized']].head())
print(f"StandardScaler Mean: {df_scaling['Age_Standardized'].mean():.2f}, Std: {df_scaling['Age_Standardized'].std():.2f}")
print(f"MinMaxScaler Min: {df_scaling['Age_Normalized'].min()}, Max: {df_scaling['Age_Normalized'].max()}")
```

**Difference between Scaling methods:**
- **Min-Max Scaling:** Scales data to a fixed range (usually 0 to 1). It is sensitive to outliers because one extreme value can compress all other data points into a tiny range.
- **StandardScaler:** Rescales data to have a mean of 0 and a standard deviation of 1 (Z-score). It does not bound values to a specific range and is generally more robust to outliers in many machine learning algorithms like SVM or Linear Regression.

**Impact of Label Encoding:**
Label Encoding converts categorical text into integers.
- **Impact:** It provides a numerical representation that models can process. However, it introduces an implicit **ordinal relationship** (e.g., 1 > 0). If the categories are nominal (like 'Male' vs 'Female'), some models might incorrectly assume 'Male' is "higher" than 'Female'.
